```python
vetor=[]
maiores=0
menores=0
iguais=0

for i in range(10):
  numero=int(input(f"Adicione o {i+1}Âº inteiro: "))
  vetor.append(numero)


for i in range(1,10):
  if vetor[0]== vetor[i]:
    iguais+=1
  elif vetor[0]< vetor[i]:
    maiores+=1
  elif vetor[0] > vetor[i]:
    menores+=1

print(f'Existem {maiores} valores maiores que o 1Âº valor inserido')
print(f'Existem {iguais} valores iguais que o 1Âº valor inserido')
print(f'Existem {menores} valores menores que o 1Âº valor inserido')
```

```python
def validar_ip(ip):
    partes = ip.split('.')
    if len(partes) != 4:
        return False

    for parte in partes:
        if not parte.isdigit() or not (0 <= int(parte) <= 255):
            return False

    return True

def ler_ips(arquivo):
    with open(arquivo, 'r') as f:
        ips = f.readlines()

    return [ip.strip() for ip in ips]

def verificar_ips(arquivo):
    ips = ler_ips(arquivo)
    for ip in ips:
        if validar_ip(ip):
            print(f"{ip} - VÃ¡lido")
        else:
            print(f"{ip} - InvÃ¡lido")


arquivo_ips = "enderecos_ips.txt"

verificar_ips(arquivo_ips)
```

```python
n_apts=int(input("Quantos apartamentos foram vendidos?"))
areas=[]

for i in range(n_apts):
  area=float(input("Qual a area do {i+1}Âº apartamento vendido ?"))
  areas.append(area)

soma_areas=sum(areas)
max_apt=max(areas)
min_apt=min(areas)

print(f'A soma das areas dos {n_apts} apartamentos Ã© de : {soma_areas}')
print(f'A area do maior apartamento Ã© : {max_apt}')
print(f'A area do menor apartamento Ã© : {min_apt}')
```

```python
def salvar_nomes(arquivo):
    with open(arquivo, 'w') as f:
        for i in range(6):
            nome = input(f"Digite o nome e sobrenome da pessoa {i + 1}: ")
            f.write(nome + '\n')

def modificar_nomes(arquivo):
    with open(arquivo, 'r') as f:
        nomes = f.readlines()

    for i in range(len(nomes)):
        nome_atual = nomes[i].strip()
        print(f"Nome atual: {nome_atual}")
        resposta = input("Deseja alterar este nome? (Sim/NÃ£o): ").strip().lower()

        if resposta == 'sim':
            novo_nome = input("Digite o novo nome e sobrenome: ")
            nomes[i] = novo_nome + '\n'

    with open(arquivo, 'w') as f:
        f.writelines(nomes)

arquivo_nomes = "nomes.txt"
salvar_nomes(arquivo_nomes)
modificar_nomes(arquivo_nomes)
```
