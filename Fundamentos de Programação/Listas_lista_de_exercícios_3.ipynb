## **MÃ©todos para listas**:

* **Contagem:**

  * **count(x)** : Retorna a quantidade de determinado elemento "x" na lista. *

* **InserÃ§Ã£o:**

  * **append(x):** Adiciona o elemento "x" ao final da lista. *

  * **extend([x,y,..,z]):** Estende a lista adicionando todos os itens do iterable ao final da lista. *

  * **insert(i,x):** Insere o elemento "x" na posiÃ§Ã£o 'i' da lista.


* **RemoÃ§Ã£o:**

  * **remove(x):** Remove o **primeiro item da lista** cujo valor Ã© igual a "x". Levanta um **ValueError** se o item nÃ£o estiver presente. *

  * **pop([i]):** Remove e retorna o item na posiÃ§Ã£o 'i' da lista. *

  * **pop():** Remove e retorna o Ãºltimo item da lista.

  * **clear():** Remove todos os itens da lista. *

## CompreensÃ£o de listas
List comprehension Ã© uma forma de gerar uma nova lista a partir de uma lista existente, aplicando uma fÃ³rmula ou condiÃ§Ã£o a cada item.

**Ex:**
```
numeros = [1, 2, 3, 4, 5, 6, 7, 8, 9]
# Usando list comprehension para calcular o cubo dos nÃºmeros Ã­mpares
cubos_impares = [x**3 for x in numeros if x % 2 != 0]
print(cubos_impares)
# resultado [1, 27, 125, 343, 729]
```



````python
lista=[]

for i in range(10):
  numero=float(input(f"Escreva o {i+1}Âº nÃºmero real : "))
  lista.append(numero)

negativos=[x for x in lista if x < 0 ]
positivos =[x for x in lista if x > 0]

print(f"\nA quantidade de nÃºmeros negativos Ã© {len(negativos)}")
print(f"A soma dos numero positivos Ã© de {sum(positivos)}")
````
