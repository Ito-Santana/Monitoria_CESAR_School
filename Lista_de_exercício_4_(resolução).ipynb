```python
armazenador=[]
nova_lista=[]

for i in range(10):
  numero=int(input('Digite o nmero : '))
  armazenador.append(numero)

l_inferior=int(input('Digite o Limite inferior : '))
l_superior=int(input('Digite o Limite superior : '))

def verificador(lista,limite_inferior,limite_superior):
  for x in lista:
    if limite_inferior <= x <= limite_superior:
      nova_lista.append(x)
  print(f'Nova lista: {sorted(nova_lista)}')

verificador(armazenador,l_inferior,l_superior)
```

```python
lista=[]

while True:
  numero=int(input("Digite o numero: "))
  if numero ==0:
    break
  else:
    lista.append(numero)

def funct(lista):
  for x in lista:
    i=0

    while i<=x:
      print(i, end=' ')
      i+=1
    print()
funct(lista)
```

```python
def minutos(valor):
  x=valor*60
  return print(x)
def segundos(valor):
  x=valor*3600
  return print(x)


horas=int(input('Digite o numero de horas :'))

print('digite 1 se quiser ver o valor em muinutos\n Digite 2 se quiser ver o valor em segundos ')

ver=int(input(" "))

if ver==1:
  minutos(horas)
elif ver==2:
  segundos(horas)

else :
  'Digite um numero correto'
```
